```python
import matplotlib.pyplot as plt
plt.style.use("fivethirtyeight")

# Change style to white background
plt.style.use('seaborn-v0_8-white')

# Data for the splits
years = range(1980, 2025)
splits = ['Entraînement', 'Validation', 'Test']
colors = ['lightblue', 'teal', 'grey']

# Plotting
fig, ax = plt.subplots(figsize=(12, 4))
ax.set_facecolor('white')
ax.hlines(1, 1980, 2016, colors=colors[0], lw=30, label=splits[0])
ax.hlines(1, 2016, 2023, colors=colors[1], lw=30, label=splits[1])
ax.hlines(1, 2023, 2024, colors=colors[2], lw=30, label=splits[2])

# Labels and customization
ax.set_xticks(years[::4])
ax.set_yticks([])
ax.set_xlabel('Date')
ax.set_title('Sous-ensembles de données', fontweight="bold", fontsize=24)
ax.legend(loc='upper center', ncol=3)
ax.grid(axis='x', linestyle='--', alpha=0.5)

# Show plot
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.dates as mdates
from datetime import datetime, timedelta

# plt.style.use("fivethirtyeight")
plt.style.use('seaborn-v0_8-white')

# Number of samples and cross-validation splits
n_samples = 10
n_splits = 9

# Generate date range from 1980 to 2016
start_date = datetime(1980, 1, 1)
end_date = datetime(2016, 1, 1)
date_range = [start_date + (end_date - start_date) * i / n_samples for i in range(n_samples)]

# Create the figure and axis
fig, ax = plt.subplots(figsize=(10, 6))

# Generate CV splits
for i in range(n_splits):
    train_indices = np.arange(0, n_samples - (i + 1))
    test_indices = np.arange(n_samples - (i + 1), n_samples - i)

    # Plot training set
    ax.barh(i + 1, len(train_indices), left=0, color='teal', label='Ensemble d\'entraînement' if i == 0 else "")

    # Plot testing set
    ax.barh(i + 1, len(test_indices), left=len(train_indices), color='grey', label='Ensemble de validation' if i == 0 else "")

# Add labels and legend
ax.set_xlabel("Date")
ax.set_ylabel("Itération de validation croisée")
ax.set_xlim(0, n_samples)
ax.set_ylim(0, n_splits + 1)
ax.set_yticks(range(1, n_splits + 1))

# Set custom x-axis date labels
ax.set_xticks(np.arange(n_samples))
ax.set_xticklabels([date.strftime('%Y') for date in date_range])
ax.legend(loc='upper right')

ax.set_title("Validation croisée avec origine de prévision glissante", fontsize=24, fontweight='bold')

# Tighten the layout
plt.tight_layout()

# Display the plot
plt.show()
```
